## 7. Load Train Model From Checkpoint

```python
# Load pipeline config and build a detection model
configs = config_util.get_configs_from_pipeline_file(files['PIPELINE_CONFIG'])
detection_model = model_builder.build(model_config=configs['model'], is_training=False)

# Restore checkpoint
ckpt = tf.compat.v2.train.Checkpoint(model=detection_model)
ckpt.restore(os.path.join(paths['CHECKPOINT_PATH'], 'ckpt-11')).expect_partial()

@tf.function
def detect_fn(image): 
    image, shapes = detection_model.preprocess(image)
    prediction_dict = detection_model.predict(image, shapes)
    detections = detection_model.postprocess(prediction_dict, shapes)
    return detections
```

## OCR Filtering 

```python
region_threshold = 0.05
```

```python
# OCR filter function
def filter_text(region, ocr_result, region_threshold):
    rectangle_size = region.shape[0]*region.shape[1]
    
    plate = [] 
    for result in ocr_result:
        length = np.sum(np.subtract(result[0][1], result[0][0]))
        height = np.sum(np.subtract(result[0][2], result[0][1]))
        
        if length*height / rectangle_size > region_threshold:
            plate.append(result[1])
    return plate
```

```python
# Calling function
filter_text(region, ocr_result, region_threshold)
```

## Combining Above Functions

#### ('Apply OCR to dectection' function and 'OCR filtering' function combined)

```python
# OCR function
def ocr_it(image, detections, detection_threshold, region_threshold):
    
    # Scores, boxes and classes above threhold
    scores = list(filter(lambda x: x> detection_threshold, detections['detection_scores']))
    boxes = detections['detection_boxes'][:len(scores)]
    classes = detections['detection_classes'][:len(scores)]
    
    # Full image dimensions
    width = image.shape[1]
    height = image.shape[0]
    
    # Apply ROI filtering and OCR
    for idx, box in enumerate(boxes):
        roi = box*[height, width, height, width]
        region = image[int(roi[0]):int(roi[2]),int(roi[1]):int(roi[3])]
        reader = easyocr.Reader(['en'])
        ocr_result = reader.readtext(region)
        
        text = filter_text(region, ocr_result, region_threshold)
        
        plt.imshow(cv2.cvtColor(region, cv2.COLOR_BGR2RGB))
        plt.show()
        print(text)
        return text, region
```

```python
# Calling function
text, region = ocr_it(image, detections, detection_threshold, region_threshold)
```

## Saving Results 

```python
# Save results function
def save_results(text, region, csv_filename, folder_path):
    img_name = '{}.jpg'.format(uuid.uuid1())
    
    cv2.imwrite(os.path.join(folder_path, img_name), region)
    
    with open(csv_filename, mode='a', newline='') as f:
        # creating a csv writer object
        csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL) 
        # converting list to string 
        string = ''.join([str(item) for item in text])
        # writing the data row
        csv_writer.writerow([img_name, string, recorded_date])
```

```python
# Calling function
save_results(text, region, 'detection_results.csv', 'Detection_images')

# Assigning OCR detection to variable 
detected_plate = ''.join([str(item) for item in text])
```

## 9. Real Time Detections from your Webcam

```python
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while cap.isOpened(): 
    ret, frame = cap.read()
    image_np = np.array(frame)
    
    input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)
    detections = detect_fn(input_tensor)
    
    num_detections = int(detections.pop('num_detections'))
    detections = {key: value[0, :num_detections].numpy()
                  for key, value in detections.items()}
    detections['num_detections'] = num_detections

    # detection_classes should be ints.
    detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

    label_id_offset = 1
    image_np_with_detections = image_np.copy()

    viz_utils.visualize_boxes_and_labels_on_image_array(
                image_np_with_detections,
                detections['detection_boxes'],
                detections['detection_classes']+label_id_offset,
                detections['detection_scores'],
                category_index,
                use_normalized_coordinates=True,
                max_boxes_to_draw=5,
                min_score_thresh=.8,
                agnostic_mode=False)
    
    try: 
        text, region = ocr_it(image_np_with_detections, detections, detection_threshold, region_threshold)
        save_results(text, region, 'detection_results.csv', 'Detection_images')
        detected_plate = ''.join([str(item) for item in text])
    except:
        pass

    cv2.imshow('object detection',  cv2.resize(image_np_with_detections, (800, 600)))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```

## 10. Confirm if the vehicle is allowed access, or if the vehicle is marked as stolen

```python
# Send email function
def send_email():

    port = 465  # For SSL
    smtp_server = 'smtp.mail.yahoo.com'   # For Gmail smtp.gmail.com
    sender_email = 'insert sender email here'  # Enter your address
    receiver_email = ['info@zonketech.co.za', 'KZN.ServiceComplaints@saps.gov.za']  # Enter receiver address
    password = input('\nType your password and press enter: ')
    message = """\
    Subject: ALERT!!!
    
    ALERT! A stolen vehicle has be deteced at the premises."""

    context = ssl.create_default_context()
    with smtplib.SMTP_SSL(smtp_server, port, context=context) as server:
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, message)
```

```python
# List of allowed vehicles
allowed_vehicles = ['FSH439EC', 'ZANDELEGP', 'FT 99 ZNGP', 'CAA68763']
```

```python
# List of stolen vehicles
stolen_vehicles = ['FH 44 BF ~ GP', 'NT798']
```

```python
# Verify number plate function
def verify_plate():
    if detected_plate in allowed_vehicles:
        print(f'Access Granted! - The vehicle entered the premises at {recorded_date}.')
        
    elif detected_plate in stolen_vehicles:
        print(f'ALERT!!! The Vehicle is Stolen - Report to Head Office and Notify the Police! - Last seen at {recorded_date}.')
        send_email()
        
    elif detected_plate not in allowed_vehicles and stolen_vehicles:
        print('Access Denied! - The vehicle does not have clearance to enter the premises.')
```

```python
# while loop for menu
while True:
    user = input("""\nWelcome! to your Automatic Number Plate Recognition System. 
    
What would you like to do:  
				V  -  Verify if vehicle is allowed or stolen
				E  -  Exit
				: """).casefold()
    
    if user == "v":
        verify_plate()
    elif user == "e":
        print("Goodbye!")
        break
    else:
        print("You have made a wrong choice, Please Try again.")
```
